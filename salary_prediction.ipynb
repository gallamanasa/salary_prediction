```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
data=pd.read_csv("Salary_dataset.csv")
```

```python
print(data.head())
```

```python
x=data[['YearsExperience']]
y=data['Salary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred_test=model.predict(x_test)
print("R2 score:", round(r2_score(y_test,y_pred_test),3))
```

```python
Years=float(input("Enter Years of Experience: "))
predicted_salary= model.predict([[Years]])
print("Predicted salary: ", round(predicted_salary[0],2))
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x))
plt.scatter(Years, predicted_salary[0])

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary prediction using LinearRegression")

plt.show()
```
